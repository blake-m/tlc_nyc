```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
card_credit_doy = pd.read_csv(r'C:\Users\blaze\Desktop\TLC_NYC\result_csv_for_tableau\payments_all_by_day_of_year.csv')
```

```python
card_credit_doy.head()
```

```python
card_credit_doy_cash = card_credit_doy[card_credit_doy['payment_type'] == 2]
card_credit_doy_card = card_credit_doy[card_credit_doy['payment_type'] == 1]
```

```python
plt.figure(figsize=(15, 25))
plt.subplot(2, 1, 1)
x_cash_count = card_credit_doy_cash['day_of_year']
y_cash_count = card_credit_doy_cash['count']
plt.plot(x_cash_count, y_cash_count, label='Cash', ls='dashed', lw=0.5, color='orange')
plt.xlabel('Day of Year')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions by Day of Year')

plt.gca()
x_card_count = card_credit_doy_card['day_of_year']
y_card_count = card_credit_doy_card['count']
plt.plot(x_card_count, y_card_count, label='Card', ls='dashed', lw=0.5, color='blue')


plt.gca()
x_card_count_rolling = card_credit_doy_card['day_of_year']
y_card_count_rolling = card_credit_doy_card['count'].rolling(5).mean()
plt.plot(x_card_count_rolling, y_card_count_rolling, label='Card - rolling mean', color='blue')

plt.gca()
x_cash_count_rolling = card_credit_doy_cash['day_of_year']
y_cash_count_rolling = card_credit_doy_cash['count'].rolling(5).mean()
plt.plot(x_cash_count_rolling, y_cash_count_rolling, label='Cash - rolling mean', color='orange')

plt.legend()



plt.subplot(2, 1, 2)
x_cash_count = card_credit_doy_cash['day_of_year']
y_cash_count = card_credit_doy_cash['sum']
plt.plot(x_cash_count, y_cash_count, label='Cash', ls='dashed', lw=0.5, color='orange')
plt.xlabel('Day of Year')
plt.ylabel('Amount in Dollars')
plt.title('Amount in Dollars by Day of Year')

plt.gca()
x_card_count = card_credit_doy_card['day_of_year']
y_card_count = card_credit_doy_card['sum']
plt.plot(x_card_count, y_card_count, label='Card', ls='dashed', lw=0.5, color='blue')


plt.gca()
x_card_count_rolling = card_credit_doy_card['day_of_year']
y_card_count_rolling = card_credit_doy_card['sum'].rolling(7).mean()
plt.plot(x_card_count_rolling, y_card_count_rolling, label='Card - rolling mean', color='blue')

plt.gca()
x_cash_count_rolling = card_credit_doy_cash['day_of_year']
y_cash_count_rolling = card_credit_doy_cash['sum'].rolling(7).mean()
plt.plot(x_cash_count_rolling, y_cash_count_rolling, label='Cash - rolling mean', color='orange')
;
```
