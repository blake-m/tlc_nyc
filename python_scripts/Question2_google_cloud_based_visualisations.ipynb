```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
card_credit_month = pd.read_csv(r"C:\Users\blaze\Desktop\Python Files\tlc_nyc\results_from_google_cloud\card_cash_transactions_by_month.csv")
# doy = day of year
card_credit_doy = pd.read_csv(r"C:\Users\blaze\Desktop\Python Files\tlc_nyc\results_from_google_cloud\card_cash_transactions_by_day_of_year.csv")
```

```python
card_credit_month.head()
```

```python
card_credit_doy.head()
```

```python
card_credit_month_cash = card_credit_month[card_credit_month['payment_type'] == 2]
card_credit_month_card = card_credit_month[card_credit_month['payment_type'] == 1]
card_credit_doy_cash = card_credit_doy[card_credit_doy['payment_type'] == 2]
card_credit_doy_card = card_credit_doy[card_credit_doy['payment_type'] == 1]
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
x_card_count = card_credit_month_card['month_']
y_card_count = card_credit_month_card['count_of_transactions']
plt.bar(x_card_count, y_card_count, label='Card', width=0.9)
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Transactions and Amount in Dollars by Month')

plt.gca()
x_cash_count = card_credit_month_cash['month_']
y_cash_count = card_credit_month_cash['count_of_transactions'].values
plt.bar(x_cash_count, y_cash_count, label='Cash')
plt.legend()


plt.subplot(2, 1, 2)

x_card_count = card_credit_month_card['month_']
y_card_count = card_credit_month_card['transaction_amounts_sum']
plt.bar(x_card_count, y_card_count, label='Card', width=0.9)
plt.xlabel('Month')
plt.ylabel('Amount in Dollars')

plt.gca()
x_cash_count = card_credit_month_cash['month_']
y_cash_count = card_credit_month_cash['transaction_amounts_sum']
plt.bar(x_cash_count, y_cash_count, label='Card')
plt.legend()
;
```

```python
plt.figure(figsize=(15, 25))
plt.subplot(2, 1, 1)
x_cash_count = card_credit_doy_cash['day_of_year']
y_cash_count = card_credit_doy_cash['count_of_transactions']
plt.plot(x_cash_count, y_cash_count, label='Cash', ls='dashed', lw=0.5, color='orange')
plt.xlabel('Day of Year')
plt.ylabel('Number of Transactions')
plt.title('Number of Transactions by Day of Year')

plt.gca()
x_card_count = card_credit_doy_card['day_of_year']
y_card_count = card_credit_doy_card['count_of_transactions']
plt.plot(x_card_count, y_card_count, label='Card', ls='dashed', lw=0.5, color='blue')


plt.gca()
x_card_count_rolling = card_credit_doy_card['day_of_year']
y_card_count_rolling = card_credit_doy_card['count_of_transactions'].rolling(7).mean()
plt.plot(x_card_count_rolling, y_card_count_rolling, label='Card - rolling mean', color='blue')

plt.gca()
x_cash_count_rolling = card_credit_doy_cash['day_of_year']
y_cash_count_rolling = card_credit_doy_cash['count_of_transactions'].rolling(7).mean()
plt.plot(x_cash_count_rolling, y_cash_count_rolling, label='Cash - rolling mean', color='orange')

plt.legend()



plt.subplot(2, 1, 2)
x_cash_count = card_credit_doy_cash['day_of_year']
y_cash_count = card_credit_doy_cash['transaction_amounts_sum']
plt.plot(x_cash_count, y_cash_count, label='Cash', ls='dashed', lw=0.5, color='orange')
plt.xlabel('Day of Year')
plt.ylabel('Amount in Dollars')
plt.title('Amount in Dollars by Day of Year')

plt.gca()
x_card_count = card_credit_doy_card['day_of_year']
y_card_count = card_credit_doy_card['transaction_amounts_sum']
plt.plot(x_card_count, y_card_count, label='Card', ls='dashed', lw=0.5, color='blue')


plt.gca()
x_card_count_rolling = card_credit_doy_card['day_of_year']
y_card_count_rolling = card_credit_doy_card['transaction_amounts_sum'].rolling(7).mean()
plt.plot(x_card_count_rolling, y_card_count_rolling, label='Card - rolling mean', color='blue')

plt.gca()
x_cash_count_rolling = card_credit_doy_cash['day_of_year']
y_cash_count_rolling = card_credit_doy_cash['transaction_amounts_sum'].rolling(7).mean()
plt.plot(x_cash_count_rolling, y_cash_count_rolling, label='Cash - rolling mean', color='orange')
;
```
